```python
# Packages for Data Analysis
import pandas as pd
import numpy as np
# from sklearn import load_boston
from sklearn import linear_model  # this model can not be used when the relationship between independent and dependent variables are not linear
from sklearn.model_selection import train_test_split # function for splitting data into traning and tests sets
```

```python
#load the boston housing dataset
from sklearn.datasets import load_boston
boston_housing = load_boston()
# print(boston_housing)
```

```python
# Dataset is messy, transform into a dataframe
#data = independent variables (x variables)
#features = column names of data or labels
#target = dependant variables or y values or price of houses

df_x = pd.DataFrame(boston_housing.data, columns=boston_housing.feature_names)
df_y = pd.DataFrame(boston_housing.target)
```

```python
#Stats on dataset (eg.mean,std) for analysis
df_x.describe()
```

```python
#Initialize linear regression model
li_reg = linear_model.LinearRegression()
```

```python
#Split data into 70 , 30
x_train, x_test, y_train, y_test = train_test_split(df_x,df_y,test_size=0.30, random_state = 40)
```

```python
#Train the model with training data
li_reg.fit(x_train,y_train)
```

```python
#Print the coefficients of model columns
print(li_reg.coef_)
```

```python
#print predictions on test data
y_pred = li_reg.predict(x_test)
print(y_pred)
```

```python
print(y_test)
```

```python
# for model performance and accuracy usimg mse
print(np.mean((y_pred-y_test)**2))
```

```python
# for model performance and accuracy usimg mse and sklearn metrics
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred))
```
